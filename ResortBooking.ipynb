-Its relatively Uniform but it's still better to avoid using the mean for dates. Here's why:

**Why not mean for dates?**

The mean of datetime gives us a timestamp that might fall between real days, especially if there's any time component. It could be weird like "2022-06-14 12:34:56".

It’s less intuitive and less robust than median for temporal data.

**Better options:**

Median — clean and robust, already aligns well with this distribution.

```python
df['reservation_date'].fillna(df['reservation_date'].median(),inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Booking_lead_time'])
plt.show()
```

**median works for Booking_lead_time:**

- The Histogram shows a highly right-skewed distribution.

- There are many small values (close to 0) and a long tail of large values.

- Mean would be dragged up by outliers (those huge lead times), which can distort imputation.
so that we can impute with Median


```python
df['Booking_lead_time'].fillna(df['Booking_lead_time'].median(),inplace=True)
```

#SCALING
- for scaling exclude categorical features,Binary features and Date format features

- As we are using various algorithms like
    1. **Logistic Regression**,
    2. **Decision Tree for cancellations**
    3. **Time-series analysis for booking trends**  
    4. **Random Forest, XGBoost for model improvement**

- its better to create scaled dataframe and unscaled dataframe

1. Logistic Regression
- Needs scaling — coefficients are sensitive to feature magnitude.

- Use: StandardScaler or RobustScaler --> we shall use Robust because some outliers are still present

2. Decision Tree
-  Does not need scaling — trees split based on feature thresholds, not distances or magnitudes.

3. Random Forest

- No scaling needed — same as Decision Tree.

4. XGBoost

- Doesn’t require scaling, but Scaling can help if features have very different magnitudes.

- Especially helpful when regularization is involved.

- Optional: StandardScaler or RobustScaler if using along with other models that do need scaling.

5. Time-Series Analysis

- May need scaling (depending on the model you use, like ARIMA doesn’t need, but LSTM does).

```python
#Rolling Statistics
df.set_index('booking_date',inplace=True)#index as booking_date
monthly_booking=df.resample('M').size()# we have to convert booking_date to monthly bookings which means count of monthly bookings per day.
rolling_mean=monthly_booking.rolling(window=12).mean()
rolling_std=monthly_booking.rolling(window=12).std()
plt.figure(figsize=(12,6))
plt.plot(monthly_booking,color='blue',label='Original')
plt.plot(rolling_mean,color='red',label='Rolling Mean')
plt.plot(rolling_std,color='green',label='Rolling Std')
```

**Not Stationary**

- The rolling mean (red) shows an upward trend over time → the mean is not constant.

- The rolling std (green) fluctuates but is relatively low and more stable, which is good, but not sufficient on its own.

- The original series (blue) has visible trends and seasonality patterns (e.g., peaks and troughs).

```python
#Dickey - Fuller Test (ADF -Test)
from statsmodels.tsa.stattools import adfuller
result=adfuller(monthly_booking)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

- p_value > 0.05 , we can reject null hypothesis and accept alternate hypothesis which means time series data is non-stationary

# Making it Stationary

1. Differencing - monthly_booking.diff().dropna()
2. log transformations
3. Detrending
4. Seasonal decomposition

```python
#removes trend
diff=monthly_booking.diff().dropna()
```

```python
#Rolling Statistics
rolling_mean=diff.rolling(window=12).mean()
rolling_std=diff.rolling(window=12).std()
plt.figure(figsize=(12,6))
plt.plot(diff,color='blue',label='Original')
plt.plot(rolling_mean,color='red',label='Rolling Mean')
plt.plot(rolling_std,color='green',label='Rolling Std')
```

- from this we can understand that mean and std deviation became constant. which means time series data became stationary

- p_value<0.05 , data became stationary

#PLOT ACF/PACF

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(sea_diff)
plot_pacf(sea_diff)
plt.show()
```

**AUTOCORRELATION FUNCTION**

- A sharp drop after lag 1

- Only a slight spike at lag 6

- Most lags after 1 are within the confidence bounds

**PARTIAL AUTOCORRELATION FUNCTION**

- PACF plot shows a gradual decline with multiple significant lags (up to around lag 5)

- not a sharp cutoff.

- This suggests a higher-order AR (p) model.



**FINDINGS**

1. ACF: Sharp drop after lag 1 ---> q = 1

2. PACF: Gradual decline up to lag 5 ---> p = 5

3. d=2 ---> did 2 times differencing the original time series to make it stationary

#MODEL BUILDING

- Through differencing i come to know that our data can become stationary

- So the best model for time series data that become stationary is ARIMA

- parameters for ARIMA also we found - AR (p): 5 , MA (q) : 1 , I(d) : 2

- we have to Use monthly_booking, not differenced one (sea_diff), since ARIMA will handle the differencing internally.

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(monthly_booking,order=(5,2,1))
model_fit=model.fit()
print(model_fit.summary())
```

**Coefficients Table**

1. AUTO REGRESSIVE LAGS (ar.L1 to ar.L5) because p=5

- p_value(ar.L) <0.05 ---> significant lag , in our model only ar.L4 has 0.002 as p_value which is less than 0.05

2. MOVING AVERAGE(ma.L1) because q=1

- p_value(ma.L)<0.05 ---> significant lag , in our model p_value for ma.L1 = 0.002

3. RESIDUALS (sigma2)

- p_value(sigma2<0.05) ---> significant lag

**Model Fitness Metrics**

1. Log Likelihood                -194.441 ----> Higher = better model fit

2. AIC                            402.883 ----> Lower = better model

3. BIC                            411.129 ----->Lower = better model

- we can compare these with other models lower can be choosen as better model

**Residual Diagnostics**

1. Ljung-Box Q (Prob Q) - Tests autocorrelation in residuals.
if p > 0.05 --> residuals are OKAY --> Better model

     - Prob(Q) = 0.69 (better)

2. Jarque-Bera (JB) - Tests normality. If p < 0.05, residuals are not normally distributed ---> Not Good
  
    - Prob(JB) = 0.00 (residuals not normal)

3. Heteroskedasticity (H) - Tests constant variance. If p < 0.05, variance is not constant ---> Not Good

    - Prob(H) = 0.02 (variance not constant)

#EXPECTED OCCUPANCY PEAKS OR DIP
- we have to find local maxima(peaks) and local minima(dips) from forecasted values

```python
from scipy.signal import argrelextrema
forecast['Occupancy'] = forecast['Occupancy'].round(2)
# Find local maxima and minima
n = 3  # sensitivity
peaks = argrelextrema(forecast['Occupancy'].values, np.greater_equal, order=n)[0]#returns index numbers as list of greater values in occupancy
dips = argrelextrema(forecast['Occupancy'].values, np.less_equal, order=n)[0]#returns index numbers as list of lesser values in occupancy

peak_dates = forecast.iloc[peaks][['Date', 'Occupancy']]#this will return dates and occupancy in peaks index
dip_dates = forecast.iloc[dips][['Date', 'Occupancy']]#this will return dates and occupancy in dips index

peak_dates.sort_values(by='Occupancy',ascending=False,inplace=True)#based on occupancy sorting is done
dip_dates.sort_values(by='Occupancy',ascending=True,inplace=True)
print("Peaks:\n", peak_dates)
print('\n')
print("Dips:\n", dip_dates)
```

#Periods of low booking (useful for promotions)

```python
#Define low threshold
low_threshold = forecast['Occupancy'].quantile(0.25)  # bottom 25% of forecasted occupancy

#Find low booking periods
low_booking_periods = forecast[(forecast['Date'] > ts['ds'].max()) & (forecast['Occupancy'] < low_threshold)]#will return future dates(forecasted), dates greater than maximum timeseries dates and occupancy less than threshold, this overall return future dates with low occupancy

#Display them
print("Low booking periods for potential promotions:")
print(low_booking_periods[['Date', 'Occupancy']])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler, OneHotEncoder, FunctionTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from prophet import Prophet
from sklearn.metrics import accuracy_score,classification_report
import pandas as pd
import numpy as np
import joblib


# Corrected transformer for yes/no column
def convert_yes_no(df):
    return df.applymap(lambda x: {'Yes': 1, 'No': 0}.get(x, 0))

# Corrected transformer for country frequency
def add_country_freq(df):
    return df.applymap(lambda x: country_freq_map.get(x, 0))



df=pd.read_excel('/content/resortbookings.xlsx')
# Define feature columns (raw)

df.rename(columns={'entry_year':'booking_year'},inplace=True)
df.rename(columns={'entry_month':'booking_month'},inplace=True)
df.rename(columns={'entry_month.1':'booking_day'},inplace=True)

# Map month names to numbers
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}


# Convert month names to numbers
df['booking_month'] = df['booking_month'].map(month_map)

peak_months=[8,7,5,10,3,4]
df['is_peak_season']=df['booking_month'].apply(lambda x: 1 if x in peak_months else 0)



# Safely create date, invalid ones become NaT (Not a Time)
df['booking_date'] = pd.to_datetime({
    'year': df['booking_year'],
    'month': df['booking_month'],
    'day': df['booking_day']
}, errors='coerce')

df = df.copy()

# Remove or fix invalid dates
invalid_rows = df['booking_date'].isna().sum()
if invalid_rows > 0:
    print(f"Found {invalid_rows} invalid booking dates. Removing them.")
    df = df.dropna(subset=['booking_date'])

# Then compute lead time
df['reservation_date'] = pd.to_datetime(df['reservation_date'], errors='coerce')
df['booking_lead_time'] = (df['booking_date'] - df['reservation_date']).dt.days
df['booking_lead_time'] = df['booking_lead_time'].clip(lower=0)

df['booking_week']=df['booking_date'].dt.isocalendar().week

feature_cols = [
    'adults', 'newborn', 'kids', 'booking_lead_time','booking_month','booking_week',
    'booking_day', 'booking_year', 'total_time', 'total_of_special_requests',
    'meal', 'reserved_room', 'resort', 'country', 'is_repeated_guest'
]
#parameters for rf model
target_col_rf = 'is_peak_season'
x_rf = df[feature_cols]
y_rf = df[target_col_rf]
# Split
x_train_rf, x_test_rf, y_train_rf, y_test_rf = train_test_split(x_rf, y_rf, test_size=0.2, random_state=42)

#parameters for xgb model
target_col_xgb ='is_canceled'
x_xgb = df[feature_cols]
y_xgb = df[target_col_xgb]
#split
x_train_xgb, x_test_xgb, y_train_xgb, y_test_xgb = train_test_split(x_xgb, y_xgb, test_size=0.2, random_state=42)


# Feature groups
numerical_features = ['adults','newborn','kids','booking_lead_time', 'total_time','total_of_special_requests']
categorical_features = ['meal', 'reserved_room', 'resort']
frequency_features = ['country']
yes_no_cols = ['is_repeated_guest']

# Fit country frequency map from training data
country_freq_map = x_train_rf['country'].value_counts().to_dict()


# Transformers
yes_no_transformer = FunctionTransformer(convert_yes_no, validate=False)
freq_encoder = FunctionTransformer(add_country_freq,validate=False)


# Preprocessors
preprocessor_scaled = ColumnTransformer(transformers=[
    ('yes_no', yes_no_transformer, yes_no_cols),
    ('num', RobustScaler(), numerical_features),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features),
    ('freq', freq_encoder, frequency_features)
])

preprocessor_unscaled = ColumnTransformer(transformers=[
    ('yes_no', yes_no_transformer, yes_no_cols),
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features),
    ('freq', freq_encoder, frequency_features)
])
#Random Forest
RF_model = RandomForestClassifier(random_state=42,class_weight='balanced')

#XGBoost
scale_pos_weight = len(y_train_xgb[y_train_xgb==0]) / len(y_train_xgb[y_train_xgb==1])
model_xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss',scale_pos_weight=scale_pos_weight,
    random_state=42)

#Prepare Prophet training data

monthly_booking = df.set_index('booking_date').resample('MS').size().asfreq('MS').fillna(0)
ts = monthly_booking.reset_index()
ts.columns = ['ds', 'y']

model_prophet=Prophet()
model_prophet.fit(ts)



# Pipelines
pipeline_rf = Pipeline([
    ('preprocessor_rf', preprocessor_scaled),
    ('model', RF_model)
])

pipeline_xgb = Pipeline([
    ('preprocessor_xgb', preprocessor_unscaled),
    ('model', model_xgb)
])

# Fit pipelines
pipeline_rf.fit(x_train_rf, y_train_rf)
pipeline_xgb.fit(x_train_xgb, y_train_xgb)

#evaluation for rf
y_pred_rf = pipeline_rf.predict(x_test_rf)
accuracy_rf = accuracy_score(y_test_rf, y_pred_rf)
print(f"Accuracy for rf: {accuracy_rf:.2f}")
class_report_rf = classification_report(y_test_rf, y_pred_rf)
print("Classification Report for rf:\n", class_report_rf)

#evaluation for xgb
y_pred_xgb = pipeline_xgb.predict(x_test_xgb)
accuracy_xgb = accuracy_score(y_test_xgb, y_pred_xgb)
print(f"Accuracy for xgb: {accuracy_xgb:.2f}")
class_report_xgb = classification_report(y_test_xgb, y_pred_xgb)
print("Classification Report for xgb:\n", class_report_xgb)
print(f"Accuracy for xgb: {accuracy_xgb:.2f}")

# Save pipelines
#joblib.dump(pipeline_rf, 'rf_pipeline.pkl')
#joblib.dump(pipeline_xgb, 'xgb_pipeline.pkl')
joblib.dump(model_prophet, 'model_prophet.pkl')
#joblib.dump(country_freq_map, 'country_freq_map.pkl')
```
